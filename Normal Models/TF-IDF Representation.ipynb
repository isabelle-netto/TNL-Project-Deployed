```python
import os
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("C://Users//koh18//Downloads//TNL Project//cleaned_training_dataset_normal.csv")

print("Dataset Shape:", df.shape)

df.head()
```

```python
print(df["sentiment"].value_counts())
```

```python
df[["comment", "clean_comment"]].head(20)
```

```python
X = df["clean_comment"]

y = df["sentiment"]
```

```python
#Train/Test Split
#Stratified Sampling
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
    
)
print("Training Samples:", len(X_train))
print("Testing Samples :", len(X_test))
```

```python
print("Training Distribution")
print(y_train.value_counts())

print("\nTesting Distribution")
print(y_test.value_counts())
```

```python
#Create TF-IDF Vectorizer
tfidf = TfidfVectorizer(
    max_features=100000,
    min_df=1,
    max_df=0.90,
    ngram_range=(1,2),
    sublinear_tf=True
)
```

```python
#Fit TF-IDF
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
print("Training Matrix:", X_train_tfidf.shape)
print("Testing Matrix :", X_test_tfidf.shape)
```

```python
print("Vocabulary Size:", len(tfidf.vocabulary_))
```

```python
feature_names = tfidf.get_feature_names_out()
print(feature_names[:50])
```

```python
os.makedirs("TF-IDF file", exist_ok=True)
joblib.dump(tfidf, "TF-IDF file/tfidf_vectorizer.pkl")
print("TF-IDF vectorizer saved.")
```

```python
os.makedirs("Feature Representations", exist_ok=True)
joblib.dump(X_train_tfidf, "Feature Representations/X_train_tfidf.pkl")
joblib.dump(X_test_tfidf, "Feature Representations/X_test_tfidf.pkl")

joblib.dump(y_train, "Feature Representations/y_train.pkl")
joblib.dump(y_test, "Feature Representations/y_test.pkl")

print("Training and testing datasets saved.")
```
